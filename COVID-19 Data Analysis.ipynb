```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import the dataset
dataset = pd.read_csv("covid19_Confirmed_dataset.csv")
dataset
```

```python
#delete the useless column
df = dataset.drop(["Lat", "Long"], axis=1, inplace=True)
```

```python
# aggregate the rows by country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
#visualize data
corona_dataset_aggregated.loc["Algeria"].plot()
corona_dataset_aggregated.loc["Austria"].plot()
corona_dataset_aggregated.loc["Afghanistan"].plot()
plt.legend()
```

```python
#calculate the first derivative of the curve
corona_dataset_aggregated.loc["Austria"].diff().plot()
```

```python
#maximum infection rate
corona_dataset_aggregated.loc["Austria"].diff().max()
```

```python
corona_dataset_aggregated.loc["Algeria"].diff().max()
```

```python
corona_dataset_aggregated.loc["Afghanistan"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

for c in countries:
    # Get the time series row for country 'c'
    country_data = corona_dataset_aggregated.loc[c]

    # Convert to numeric only (ignore text like country name, etc.)
    numeric_data = pd.to_numeric(country_data, errors='coerce')

    # Calculate daily new cases and take the max value
    max_infection_rates.append(numeric_data.diff().max())

# Add the result as a new column (align it properly)
corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates
```
